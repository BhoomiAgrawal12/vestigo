```python
import pandas as pd

df = pd.read_csv('dataset.csv', on_bad_lines='skip')
valid_labels = [
    'Non-Crypto', 'ECC', 'SHA-1', 'SHA-224', 'SHA-256',
    'PRNG', 'RSA-1024', 'RSA-4096', 'XOR-CIPHER',
    'AES-128', 'AES-192', 'AES-256'
]

clean_df = df[df['label'].isin(valid_labels)].copy()

# Format: {(filename, function_name): label}
label_map = dict(zip(zip(clean_df['filename'], clean_df['function_name']), clean_df['label']))

print(f"Cleaned dataset size: {len(clean_df)} functions")
print(f"Sample Entry: {list(label_map.items())[0]}")
```
